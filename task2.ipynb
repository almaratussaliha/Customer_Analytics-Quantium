```python
import warnings
import plotly.express as px
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
warnings.filterwarnings('ignore')
# import sweetviz
```

```python
df = pd.read_csv('dataset/data_cleaning.csv')
df.sample(5)
```

```python
df.info()
```

```python
# change date types to date time
df['DATE']=pd.to_datetime(df['DATE'])
```

```python
df.dtypes
```

```python
# get month
df['month'] = pd.DatetimeIndex(df['DATE']).month
df.sample(3)
```

```python
# get month name
dates = pd.to_datetime(df['DATE'])
df['month_name'] = dates.dt.month_name(locale='English')
df.sample(5)
```

```python
monthly_sales = df.groupby('month_name').agg({'TOT_SALES' : 'sum'}).reset_index()
monthly_sales.columns = ['month', 'tot_sales']
monthly_sales
```
